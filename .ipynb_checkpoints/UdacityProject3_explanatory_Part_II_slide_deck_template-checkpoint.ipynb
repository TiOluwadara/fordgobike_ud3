## Investigation Overview


> In this investigation, I will look at the types of users, how they share their trips and the preferred days and hours of riding and how that relates with the types of users. I will also check the age and gender distribution of the users.



## Dataset Overview

>Ford GoBike System dataset includes information about individual rides made in a bike-sharing system covering the greater San Francisco Bay area. The cleaned dataset has 174752 features and 29 columns which contains information about the duration, start time, end time of bike trips, the types of users, names of start and end stations with their ids, latitude and longitude, the id of the bikes used, birth year. age and genders of the members.  


```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
#the cleaned data set from the exploratory phase was imported

dfClean = pd.read_csv('C:/Users/HP/Documents/python_one_campus/python_workspace/fordgobike_clean.csv')
```

## Use Type Distribution
The distribution of the user types, most of the user types are subscribers

```python
#visualisation
sns.countplot(data = dfClean, x = 'user_type')
plt.title('User Type Distributuon')
plt.xlabel('user types')
plt.ylabel('distribution');
```

## Bike Share
The distribution of bike sharing, most of the users do not share their trip information

```python
#visualisation
plt.figure(figsize=(10,6))
sorted_counts = dfClean['bike_share_for_all_trip'].value_counts()

plt.pie(sorted_counts, labels = sorted_counts.index, startangle = 90,
        counterclock = False, autopct='%1.2f%%')
plt.axis('square')
plt.title('Bike Share', fontsize=15);
```

## Average User Duration
To check the average duration users(subscribers and customers) spend on their bike rides. It is shown that the customers spend more time in comparison to the subscribers

```python
#visualisation
dfClean[['user_type', 'duration_min']].groupby('user_type').mean().plot(kind = 'bar')
plt.title('Average User Duration')
plt.xlabel('user type')
plt.ylabel('duration');
```

## Users Weekday Preference
To check the preferred days of riding for each user type. Most of the users ride more on thursdays. The number of subscribers ridng on saturdays and sundays are reduced while it is increased during the week. It might be because they have to ride to work during the week.

```python
#Question to check if user type relates with start and end days of trips in any way

#visualisation
plt.figure(figsize = [16, 5])


# left plot: start day
plt.subplot(1, 2, 1)
ax1 = sns.countplot(data = dfClean, hue = 'user_type', x = 'start_weekday_name')
plt.title('Users Start Weekday Preference')

# right plot: end day
plt.subplot(1, 2, 2)
sns.countplot(data = dfClean, hue = 'user_type', x = 'end_weekday_name')
plt.ylim(ax1.get_ylim()); # set y-axis limits to be same as left plot
plt.title('Users End Weekday Preference');

```

## Users Hour Preference
To check the preferred time of riding for users. It is shown that. Most subscribers prefer early morning (8am) and evening rides (5pm) which also relates with the beginning work hours and closing work hours.

```python
#Question to check if user type relates with start and end days of trips in any way

#visualisation
plt.figure(figsize = [16, 5])

# left plot: start day
plt.subplot(1, 2, 1)
ax1 = sns.countplot(data = dfClean, hue = 'user_type', x = 'start_hour')
plt.title('Users Start Hour Preference')

# right plot: end day
plt.subplot(1, 2, 2)
sns.countplot(data = dfClean, hue = 'user_type', x = 'end_hour')
plt.ylim(ax1.get_ylim()); # set y-axis limits to be same as left plot
plt.title('Users End Hour Preference');
```

## Users Bike Sharing
To check the users that share their bike trips. The only people that shared their trips were all subscribers, although a larger percentage of them, didnt share. It might be that there is no share feature for customers, and that being a subscriber increases your chance of sharing your trip.

```python
#visualisation
sns.countplot(data = dfClean, hue = 'bike_share_for_all_trip', x = 'user_type')
plt.title('Users Bike Sharing')
plt.xlabel('user type')
plt.ylabel('count');
```

## User Type Duration
To check the duration accross weekdays each user spend. Customers prefer ridng on weekends probably because they just ride for fun and don't use it routinely, they also spend more time on their rides because they are probably not paying or they want to use all they can get on free versions before going ahead to subscribe.

```python
#visualistion
plt.figure(figsize=(10,6))
sns.barplot(data = dfClean, x= 'start_weekday_name', y='duration_min', hue='user_type')
plt.title('User Type Duration')
plt.xlabel('user_type')
plt.ylabel('duration');
```

## Peak Riding Hours
Theres is an increse in the duration spent on riding between 2am and 3am for monday, wednesday, thursday, saturday and sunday, while Tuesday peaks at 12am.
It seems people like spending more time on riding in the middle of the night compared to during the day

```python
bin_edges = np.arange(0, dfClean['start_weekday'].max()+1, 1)
g=sns.FacetGrid(data = dfClean, col='start_weekday_name', col_wrap = 2, height = 5, aspect = 1.5)
g.map(sns.pointplot,'start_hour', 'duration_min');
```

## Popular Stations
The preferred start stations constitute the end stations, which supports the insight that the users are routine riders that use the stations everyday to go to thier work places. 

```python
#visualisation
dfClean['start_station_name'].value_counts().sort_values(ascending = False).head(10).plot.barh();
plt.title('Popular Start Stations')
plt.xlabel('counts')
plt.ylabel('stations')
plt.show;
```

```python
#visualisation
dfClean['end_station_name'].value_counts().sort_values(ascending = False).head(10).plot.barh();
plt.title('Popular End Stations')
plt.xlabel('counts')
plt.ylabel('stations')
plt.show;
```

## Age Distribution
It is also noted that the average age range of riders are 30-40, because this is middle age where the energy to dispense for riding is available. About 29 people are above 100, I removed their representation in the plot. The older age group might be the leisure riders.

```python
#visualisation of age distribution

plt.figure(figsize=(10,6))
bin_edges = np.arange(21, 144+1, 1)
plt.hist(data = dfClean, x ='Age', bins = bin_edges, rwidth = 0.7)
plt.xlim((21,100)); #I am not including the outliers
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Distribution');
```

## Gender Duration
Males spend less time on their rides because they are more physically fit in comparison to females.

```python
#visualisation
dfClean[['member_gender', 'duration_min']].groupby('member_gender').mean().plot(kind = 'bar')
plt.title('Gender Duration')
plt.xlabel('Gender')
plt.ylabel('Duration');
```
